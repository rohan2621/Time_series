```python
from statsmodels.tsa.stattools import adfuller

def adf_test(timeseries):
    """
    Performs the Augmented Dickey-Fuller test for stationarity and prints the results.

    Parameters:
    timeseries (pd.Series or np.array): The time series data to test.
    """
    print("Results of Augmented Dickey-Fuller Test:")
    # The adfuller function returns a tuple of results
    adf_result = adfuller(timeseries, autolag='AIC')
    
    output = pd.Series(adf_result[0:4], index=['Test Statistic', 'p-value', 'Number of Lags Used', 'Number of Observations Used'])
    for key, value in adf_result[4].items():
        output[f'Critical Value ({key})'] = value
    print(output)
    
    # Interpretation based on p-value and test statistic
    if adf_result[1] <= 0.05:
        
        print("\nConclusion: Reject the null hypothesis (H0). The time series is likely stationary.")
    else:
        print("\nConclusion: Fail to reject the null hypothesis (H0). The time series is likely non-stationary.")
```

```python
adf_test(df["IT_solar_generation"])
```

```python
train_size=int(len(df)*.8)
train,test=(df['IT_load_new'][:train_size],df['IT_load_new'][train_size:])
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
for i in range(2, 4):
    for j in range(2, 4):
        try:
            model = ARIMA(train, order=(i, 0, j))
            model_fit = model.fit()
            predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1)
            rmse = sqrt(mean_squared_error(test, predictions))
            print(f"For p={i} and q={j}, RMSE is {rmse:.4f}")
        except Exception as e:
            print(f"Error for p={i}, q={j}: {e}")
```

```python
import matplotlib.pyplot as plt

model = ARIMA(train, order=(2,0,0))
model_fit = model.fit()
predictions = model_fit.predict(start=len(train), end=len(train)+len(test)-1)

plt.plot(test, label="Actual")
plt.plot(predictions, label="Predicted", color="red")
plt.legend()
plt.show()
```

Conclusion on ARIMA Performance

RMSE values are very high , indicating that the models predictions are far from the actual values.

Predicted values are almost constant, forming a nearly straight line along the median of the data, which shows the model is not capturing the trends or patterns in the series.

ARIMA fails here because:

Your time series may have non-linear patterns or high variance that ARIMA cannot model.

There could be seasonality, sudden jumps, or structural changes in the data.

ARIMA assumes linear relationships and stationarity; if these assumptions are violated, predictions flatten.
